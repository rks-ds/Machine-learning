```python
n=train.shape[0]
p=test.shape[0]
```

```python
#making bag of words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
desc=[]
for i in range(n):
    desc.append(train.description[i])
```

```python
for i in range(p):
    desc.append(test.description[i])
```

```python
vectorizer=CountVectorizer(analyzer="word",max_features=5000)
```

```python
data_features=vectorizer.fit_transform(desc)
```

```python
data_features=data_features.toarray()
```

```python
data_features.shape
```

```python
vocab=vectorizer.get_feature_names()
```

```python
data=pd.DataFrame(data=data_features,columns=vocab)
```
