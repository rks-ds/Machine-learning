```python
import numpy as np
col=train.columns
r,c=train.shape

n_train=pd.DataFrame(index=np.arange(0,r),columns=['Wilderness_Area','Soil_Type','Cover_Type'])

for i in range(0,r):
    w=0
    for j in range(11,15):
        if(train.iloc[i,j]==1):
            w=j-10
            break
    n_train.iloc[i,0]=w      
```

```python
pd.crosstab(n_train['Wilderness_Area'],columns='Count')
```

```python
train.columns[14:52]
```

```python
for i in range(0,r):
    s=0
    for j in range(14,53):
        if(train.iloc[i,j]==1):
            s=j-14+1
            break
    n_train.iloc[i,1]=s
```

```python
pd.crosstab(n_train['Soil_Type'],columns='Count')
#Soil_Type8 and Soil_Type25 has only one observation and that is why they are most skewed.
```

```python
n_train['Cover_Type']=train['Cover_Type']
```

```python
for i in range(0,15120):
    if(n_train.iloc[i,1]>13 and n_train.iloc[i,1]<39):
        n_train.iloc[i,1]=n_train.iloc[i,1]+2


    
```

```python
pd.crosstab(n_train['Soil_Type'],columns='Count')
```

```python
for i in range(0,15120):
    if(n_train.iloc[i,1]>6 and n_train.iloc[i,1]<14):
        n_train.iloc[i,1]=n_train.iloc[i,1]+1
```
